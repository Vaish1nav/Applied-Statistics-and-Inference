Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

Answer: In hypothesis testing, we evaluate evidence to decide whether to reject or fail to reject the null hypothesis. During this process, two kinds of errors can occur: Type I and Type II errors. A Type I error happens when we reject the null hypothesis even though it is actually true. This is also known as a false positive, meaning we detect an effect that does not actually exist. A Type II error, on the other hand, occurs when we fail to reject the null hypothesis when it is actually false. This is a false negative, meaning we miss detecting a real effect.

1. Type I Error (α):

    -Rejecting a true null hypothesis

    -False positive

    -Probability = significance level (α), usually 0.05

2. Type II Error (β):

    -Failing to reject a false null hypothesis

    -False negative

    -Probability = β

3. Impact on Decisions:

    -Lower α reduces false positives but increases false negatives

    -Balancing α and β is crucial depending on consequences (e.g., medical tests)

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Answer: The P-value is the probability of obtaining results at least as extreme as the observed data, assuming that the null hypothesis is true. It helps quantify how compatible the sample evidence is with the assumption that there is no effect or no difference. A low P-value suggests that the observed results would be very unlikely if the null hypothesis were true.

**Interpretation:**

P-value ≤ α:
There is strong evidence against the null hypothesis → we reject H₀.

P-value > α:
Evidence is insufficient to reject H₀ → we fail to reject it.

A P-value does not tell whether the null hypothesis is true or false.

It also does not measure the size or importance of an effect.


Question 3:Explain the difference between a Z-test and a T-test, including when to use each.

Answer: The Z-test and T-test are statistical tests used to compare sample data with population parameters, but they differ based on sample size and the availability of population standard deviation. A Z-test is used when the population standard deviation is known and the sample size is large. A T-test, meanwhile, is used when the population standard deviation is unknown, especially with small sample sizes. The T-test uses the t-distribution, which has heavier tails to account for extra uncertainty.

1. Z-test:

    Population standard deviation known

    Large sample size (n > 30)

    Data is normally distributed

2. T-test:

    Population standard deviation unknown

    Small sample size (n < 30)

    Data approximately normal

**Difference:**

Z-test uses Z-distribution

T-test uses Student’s t-distribution

T-test gives more reliable results for small samples

Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

Answer: A confidence interval (CI) is a range of values used to estimate a population parameter, such as the mean, with a specific level of confidence (e.g., 95%). It provides an interval that likely contains the true value of the parameter. The confidence interval is calculated using the sample mean and the margin of error. The margin of error (ME) determines how wide or narrow the interval is, which affects how precise the estimate is.

**How Margin of Error Influences the Interval**
   
1. Higher margin of error -> Wider confidence interval
      
    -Indicates less precision

    -Often caused by small sample size or high variability

2. Lower margin of error -> Narrower confidence interval

    -Indicates more precision

    -Usually achieved with large sample sizes

3. Margin of error depends on:

    -Sample size (larger sample means smaller margin)

    -Variability in data (higher SD means larger margin)

    -Confidence level chosen (higher confidence means larger margin)

Question 5:Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Answer: ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine whether at least one group mean is significantly different from the others. The purpose of ANOVA is to test for overall differences without performing multiple t-tests, which increases the risk of Type I error. ANOVA uses the F-statistic, which compares the variation between groups to the variation within groups.


**Assumptions:**

1. Independence: Each observation must be independent.

2. Normality: Each group should follow a normal distribution.

3. Equal Variances (Homogeneity): All groups must have similar variances.

**How ANOVA Extends Hypothesis Testing:**

1. Allows comparison of more than two group means simultaneously.

2. Reduces the error rate compared to multiple t-tests.

3. Identifies whether differences exist, but not which groups differ—post-hoc tests are used for that.



```python
'''Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.'''

import numpy as np
from scipy.stats import norm

# Sample data
data = np.array([12, 15, 14, 16, 13, 14, 15])
sample_mean = np.mean(data)

# Given population parameters
pop_mean = 14
pop_std = 2

# Z-test calculation
z_value = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(data)))
p_value = 1 - norm.cdf(abs(z_value))

print("Z-value:", z_value)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")


```

```python
'''Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.'''

import numpy as np
import matplotlib.pyplot as plt

# Non-normal distribution (exponential)
population = np.random.exponential(scale=2, size=100000)

sample_means = []

# Generate many sample means
for _ in range(1000):
    sample = np.random.choice(population, size=50)
    sample_means.append(np.mean(sample))

# Plot sample means
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Distribution of Sample Means (CLT Demonstration)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
'''Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_ci(data, confidence=0.95):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)

    z = norm.ppf((1 + confidence) / 2)
    margin = z * (std / np.sqrt(n))

    lower = mean - margin
    upper = mean + margin

    print(f"Mean: {mean}")
    print(f"{confidence*100}% Confidence Interval: ({lower}, {upper})")

    # Visualization
    plt.errorbar(1, mean, yerr=margin, fmt='o')
    plt.title("Confidence Interval for Sample Mean")
    plt.xlim(0, 2)
    plt.show()

# Example usage
data = np.random.randn(50)
plot_ci(data)
```

```python
'''Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.'''

import numpy as np
from scipy.stats import chisquare

# Observed and expected data
observed = np.array([25, 30, 20, 25])
expected = np.array([25, 25, 25, 25])

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Observed distribution is significantly different from expected.")
else:
    print("No significant difference between observed and expected distributions.")

```
